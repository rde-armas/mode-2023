# Golden Search

```python
import math

#Datos
datos  = [-0.47, 0.89, 0.26, -0.59, 0.37, 0.54, 0.87]
```

```python
def density(x, theta):
  return (1 + x*theta)/2

def mle(theta, x_datos):
  res  = len(x_datos) * math.log(2)
  for i in x_datos:
    res = res - math.log(1 + theta * i)
  return res

phi = (1 + math.sqrt(5)) / 2
#resphi = 2 - phi

def golden_search(f, a, b, datos, tau=1e-5):
    theta1 = a + (b-a) / phi
    theta2 = b - (b-a) / phi
    f1 = f(theta1, datos)
    f2 = f(theta2, datos)
    while abs(b - a) > tau:
      f1 = f(theta1, datos)
      f2 = f(theta2, datos)
      if f1 < f2:
        res = theta1
        a = theta2
        theta2 = theta1
        theta1 = a + (b - a) / phi
      else:
        res = theta2
        b = theta1
        theta1 = theta2
        theta2 = b - (b-a) / phi
    return res
```

```python
#intervalo -1 a 1
golden_search(mle, -1, 1, datos)
```
